```python
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Define tus tópicos

for i in topics:
    
    topic1 = i[0]
    topic2 = i[1]


    # Tokeniza tus tópicos
    tokens1 = topic1.lower().split()
    tokens2 = topic2.lower().split()

    # Entrena un modelo Word2Vec (puedes usar un modelo preentrenado si tus datos son pequeños)
    corpus = [tokens1, tokens2]
    model = Word2Vec(corpus, vector_size=100, window=5, min_count=1, workers=4)

    # Obtiene los vectores de cada tópico
    vector1 = np.mean([model.wv[word] for word in tokens1 if word in model.wv], axis=0)
    vector2 = np.mean([model.wv[word] for word in tokens2 if word in model.wv], axis=0)

    # Asegúrate de que ambos vectores tengan la misma longitud
    if vector1.shape[0] != vector2.shape[0]:
        raise ValueError("Los vectores tienen diferentes longitudes. Ajusta el tamaño del espacio vectorial.")

    # Calcula la similitud coseno entre los vectores
    similarity_score = cosine_similarity(vector1.reshape(1, -1), vector2.reshape(1, -1))[0, 0]

    print(f"Similitud entre los tópicos: {similarity_score}")
```

```python
import spacy
from sklearn.metrics.pairwise import cosine_similarity

# Cargar el modelo de spaCy (asegúrate de tenerlo instalado previamente: pip install spacy)
nlp = spacy.load("es_core_news_sm")
total = 0
# Definir la función para calcular la similitud entre dos tópicos
def calcular_similitud(t1, t2):
    # Tokenizar y procesar los textos con spaCy
    doc_t1 = nlp(t1)
    doc_t2 = nlp(t2)

    # Obtener los vectores promedio de los textos
    vec_t1 = doc_t1.vector.reshape(1, -1)
    vec_t2 = doc_t2.vector.reshape(1, -1)

    # Calcular la similitud coseno entre los dos vectores
    similitud = cosine_similarity(vec_t1, vec_t2)[0, 0]
    return similitud

for i in topics:
    
    topico_1 = i[0]
    topico_2 = i[1]


    similitud = calcular_similitud(topico_1, topico_2)
    print(f"Similitud entre '{topico_1}' y '{topico_2}': {similitud:.2f}")
    total += similitud
print(total / len(topics))
```

```python
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics.pairwise import cosine_similarity
import torch

total = 0
# Cargar el tokenizador y el modelo BERT preentrenado (puedes cambiar el modelo según tus necesidades)
modelo_bert = AutoModel.from_pretrained("dccuchile/bert-base-spanish-wwm-uncased")
tokenizador_bert = AutoTokenizer.from_pretrained("dccuchile/bert-base-spanish-wwm-uncased")

# Definir la función para calcular la similitud entre dos tópicos con BERT
def calcular_similitud_bert(t1, t2):
    # Tokenizar y obtener los embeddings de BERT para cada tópico
    tokens_t1 = tokenizador_bert(t1, return_tensors='pt')
    tokens_t2 = tokenizador_bert(t2, return_tensors='pt')

    with torch.no_grad():
        embeddings_t1 = modelo_bert(**tokens_t1)['last_hidden_state']
        embeddings_t2 = modelo_bert(**tokens_t2)['last_hidden_state']

    # Calcular la similitud coseno entre los vectores de embeddings
    similitud = cosine_similarity(embeddings_t1.mean(dim=1), embeddings_t2.mean(dim=1))[0, 0]
    return similitud

for i in topics:
    
    topico_1 = i[0]
    topico_2 = i[1] 

    similitud_bert = calcular_similitud_bert(topico_1, topico_2)
    total += similitud_bert
    print(f"Similitud BERT entre '{topico_1}' y '{topico_2}': {similitud_bert:.2f}")

print(total / len(topics))
```

```python
import transformers
import numpy as np

# Load the RoBERTa model and tokenizer
model = transformers.RobertaModel.from_pretrained('roberta-base')
tokenizer = transformers.RobertaTokenizer.from_pretrained('roberta-base')


total = 0

for i in topics:
    # Tokenizar y codificar los textos
    text1 = i[0]
    text2 = i[1]
    tokens1 = tokenizer(text1, max_length=512, truncation=True, padding=True, return_tensors='pt')
    tokens2 = tokenizer(text2, max_length=512, truncation=True, padding=True, return_tensors='pt')

    # Obtener las embeddings de los textos utilizando el modelo RoBERTa
    with torch.no_grad():
        embeddings1 = model(**tokens1).last_hidden_state.mean(dim=1)
        embeddings2 = model(**tokens2).last_hidden_state.mean(dim=1)

    # Calcular la similitud coseno entre las embeddings
    similarity = np.dot(embeddings1, embeddings2.T) / (np.linalg.norm(embeddings1) * np.linalg.norm(embeddings2))
    similarity = similarity.item()  # Convertir el resultado a un número escalar

    print(f"Similitud entre '{text1}' y '{text2}': {similarity:.2f}")

    total += similarity

print("Similitud total:", total/ len(topics))
```

```python
import transformers
import numpy as np

# Load the BERT model and tokenizer
model = transformers.BertModel.from_pretrained('bert-base-uncased')
tokenizer = transformers.BertTokenizer.from_pretrained('bert-base-uncased')

total = 0

for i in topics:
    
    # Tokenize and encode the texts
    text1 = i[0]
    text2 = i[1]

    # Tokenizar y codificar los textos
    tokens1 = tokenizer(text1, max_length=512, truncation=True, padding=True, return_tensors='pt')
    tokens2 = tokenizer(text2, max_length=512, truncation=True, padding=True, return_tensors='pt')

    # Obtener las embeddings de los textos utilizando el modelo BERT
    with torch.no_grad():
        embeddings1 = model(**tokens1).last_hidden_state.mean(dim=1)
        embeddings2 = model(**tokens2).last_hidden_state.mean(dim=1)

    # Calcular la similitud coseno entre las embeddings
    similarity = np.dot(embeddings1, embeddings2.T) / (np.linalg.norm(embeddings1) * np.linalg.norm(embeddings2))
    similarity = similarity.item()
    total += similarity
    print(f"Similitud entre '{text1}' y '{text2}': {similarity:.2f}")


print("Similitud total:", total/ len(topics))
```

```python
from transformers import AutoTokenizer, AutoModel
import torch


#Mean Pooling - Take attention mask into account for correct averaging
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)


# Sentences we want sentence embeddings for
sentences = ['Mi nombre es Siddhartha', 'Mis amigos me llamaron por mi nombre Siddhartha']

# Load model from HuggingFace Hub
tokenizer = AutoTokenizer.from_pretrained('hiiamsid/sentence_similarity_spanish_es')
model = AutoModel.from_pretrained('hiiamsid/sentence_similarity_spanish_es')

# Tokenize sentences
encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

# Compute token embeddings
with torch.no_grad():
    model_output = model(**encoded_input)

# Perform pooling. In this case, max pooling.
sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

print("Sentence embeddings:")
print(sentence_embeddings[0])
```

```python
import torch
import transformers

model = transformers.BertModel.from_pretrained('bert-base-uncased')
tokenizer = transformers.BertTokenizer.from_pretrained('bert-base-uncased')
# Tokenizar y calcular la similitud coseno entre dos textos
def cosine_similarity(text1, text2, model, tokenizer):
    # Tokenizar y codificar los textos
    tokens1 = tokenizer(text1, max_length=512, truncation=True, padding=True, return_tensors='pt')
    tokens2 = tokenizer(text2, max_length=512, truncation=True, padding=True, return_tensors='pt')

    # Obtener las embeddings de los textos utilizando el modelo
    with torch.no_grad():
        embeddings1 = model(**tokens1).last_hidden_state.mean(dim=1)
        embeddings2 = model(**tokens2).last_hidden_state.mean(dim=1)

    # Calcular la similitud coseno entre las embeddings
    cosine_similarity = torch.nn.functional.cosine_similarity(embeddings1, embeddings2, dim=1)

    return cosine_similarity.item()  # Convertir el resultado a un número escalar

total_similarity = 0

# Testear la función
for i in range(30):
    
    # Selecciona una sublista al azar
    random_sublist = random.choice(topics_a)
    random_element_1 = random.choice(random_sublist)
    
    random_sublist = random.choice(topics_a)
    random_element_2 = random.choice(random_sublist)

    text1 = random_element_1
    text2 = random_element_2

    similarity = cosine_similarity(text1, text2, model, tokenizer)

    total_similarity += similarity
        
    print(f"Similarity between '{text1}' and '{text2}': {similarity:.2f}")

#print("Total similarity:", total_similarity / len(topics))
```
